Imprimir un listado con los n√∫meros impares desde 1 hasta 999 y seguidamente otro listado con los n√∫meros pares desde 2 hasta 1000.
image.png

```python
def clasificacion(): # se nombra funci√≥n y se crean los grupos vacios
  pares = []
  impares = []
  for i in range(1,1001) : # Se crea el rango y la funcion como tal, si el residuo de una division por 2 es 0, es par, y se a√±ade al grupo de pares o al de impares respectivamente
    if i % 2 == 0 :
      pares.append(i)
    else :
      impares.append(i)

  print((f"Numeros impares: {impares}")) # los print
  print((f"Numeros pares: {pares}"))

if __name__ == "__main__": # la fiuncion principal
  clasificacion()

```

Imprimir el factorial de un n√∫mero natural n dado.

```python
def factorial():
  fact = 1 # se iguala a 1 para hacer la operacion matematica
  for i in range(1,n+1): # se crea el rango con el numero ingresado
    fact *= i #operacion
  print(f'El factorial de {n} es {fact}')

if __name__ == '__main__':  # principal
  n = int(input('Ingrese un numero: ')) #Ingreso del dato
  factorial() # llama a la funcion
```

Implementar un programa que ingrese un n√∫mero de 2 a 50 y muestre sus divisores.

```python
def divisores(num): # crear funcion
  if 2 <= num <= 50: #rango de 2 a 50
    print(f"Divisores de {num}:")
    for i in range(1, num + 1): #rango para obtener divisores
      if num % i == 0: # se obtienen divisores
        print(i) #se imprimen
  else:
    print("El n√∫mero debe estar entre 2 y 50.") # por si ponen un numero mayor a 50

if __name__ == "__main__": # funcion principal, obtencion de datos, y llamada de la funcion divisores
  num = int(input("Ingrese un n√∫mero entre 2 y 50: "))
  divisores(num)
```

Implementar el algoritmo que muestre los n√∫meros primos del 1 al 100. Nota: use funciones


```python
def es_primo(numero):# primera funcion
  #Clasificacion divisores para llegar a los primos
  if numero <= 1:
    return False
  if numero <= 3:
    return True
  if numero % 2 == 0 or numero % 3 == 0:
    return False
  i = 5
  while i * i <= numero:
    if numero % i == 0 or numero % (i + 2) == 0:
      return False
    i += 6
  return True

def primos(): # los imprime
  print("N√∫meros primos del 1 al 100:")
  for numero in range(1, 101):
    if es_primo(numero):
      print(numero)

if __name__ == "__main__":
  primos() # llama a las funciones
```
